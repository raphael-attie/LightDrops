```python
import astropy.io.fits as fits
from pathlib import Path
import glob, os
import matplotlib.pyplot as plt
import numpy as np
from astropy.stats import median_absolute_deviation as mad

def rescale(image, mref, sref, roi=None):
    imageV = image if roi is None else image[roi]
    m1 = np.nanmedian(imageV)
#     s1 = np.sqrt(bwmv(imageV))
    s1 = mad(imageV, ignore_nan=True)
    ki = sref/s1
    rimage = ki*(image - m1) + mref
    return rimage
```

```python
datadir = Path("C:\Data/FIRECAPTURE/Moon/231220/pipp_20201225_224653/Moon_180952")
datadir.exists()
```

```python
fitsf = sorted(list(datadir.glob('r_Moon*.fit')))
len(fitsf)
```

```python
sample = fits.getdata(fitsf[0])
```

```python
plt.figure(figsize=(10,4))
plt.imshow(sample, cmap='gray', origin='lower')
plt.axis('off')
plt.tight_layout()
```

```python
data = np.array([fits.getdata(f) for f in fitsf])
data.shape
```

```python
data.max(), data.min()
```

```python
data = np.array([d.astype(np.float32) for d in data if d.max() < 255])
data.shape
```

```python
tsum = data.sum(axis=-1).sum(axis=-1)
tsum.shape
plt.figure()
plt.plot(tsum)
```

```python
tmax = tsum.argmax()
tmax
```

```python
data[data == 0] = np.nan
```

```python
mref = np.nanmedian(data[tmax])
sref = np.nanstd(data[tmax])
```

```python
for i, d in enumerate(data):
    data[i] = rescale(d, mref, sref)
```

```python
data[np.isnan(data)] = 0
```

```python
data.max(), data.min()
```

```python
med_data = np.median(data, axis=0)
mean_data = np.mean(data, axis=0)
```

```python
med_data.max(), med_data.min(), mean_data.max(), mean_data.min()
```

```python
roi = np.s_[500:1000,300:700]
fig, axs = plt.subplots(nrows=1, ncols=3, figsize= (22,8))
axs[0].imshow(data[0][roi], origin='lower', cmap='gray')
axs[0].set_title('raw')
axs[0].axis('off')
axs[1].imshow(mean_data[roi], origin='lower', cmap='gray')
axs[1].set_title('mean rescaled')
axs[1].axis('off')
axs[2].imshow(med_data[roi], origin='lower', cmap='gray')
axs[2].set_title('median rescaled')
axs[2].axis('off')
plt.tight_layout()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1, figsize= (20,10))
axs.imshow(med_data, origin='lower', cmap='gray')
axs.set_title('median')
```

```python
med_data2 = med_data - med_data.min()
mean_data2 = mean_data - mean_data.min()

med_data_tiff = np.round(med_data2/med_data2.max() * 65535).astype(np.uint16)
mean_data_tiff = np.round(mean_data2/mean_data2.max() * 65535).astype(np.uint16)
mean_data2.max(), mean_data2.min()
```
